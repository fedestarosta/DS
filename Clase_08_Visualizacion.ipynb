## MATPLOTLIB VS SEABORN: ¿por qué y para qué?

+ Desventajas de Matplotlib
    + solo permite graficos simple
    + sintaxis complicada y extensa
+ Ventajas de Seaborn
    + Es una libreria construida a partir de Matplotlib
    + sintaxis intuituva y simple
    + permite visualizar graficos mas complejos 
    + Used for plotting statistical graphs, identifying trends, relationships & outliers

### RELPLOT
+ Se utiliza para observar la relación entre dos variables continuas. Por ejemplo, ¿existe una correlación entre el Nivel de Necesidades Básicas Insatisfechas (NBI) de un radio censal y el porcentaje de votos obtenido por una coalición?


```python
#Los diagramas de dispersión visualizan cada tupla de valores x e y como un solo punto y el diagrama formará una nube de puntos
sns.set(style = 'darkgrid')

# Analizamos la relación entre Participación y el Porcentaje que sacó el ganador
sns.relplot(x = 'Participacion', y = 'Porcentaje voto blanco', data = df_elecciones)

# Elegimos algunas provincias que queremos que tengan la etiqueta de texto - OPCIONAL
provincias_a_marcar = ['BUENOS AIRES', 'CORDOBA', 'SANTA FE', 'MENDOZA', 'FORMOSA']
for _, fila in df_elecciones.iterrows():
    if fila['Provincia'] in provincias_a_marcar:
        plt.text(fila['Participacion'], fila['Porcentaje voto blanco'], fila['Provincia'], fontsize=9, fontweight='bold')        
    plt.legend(bbox_to_anchor=(1.02, 1))
```

```python
sns.set(style = 'darkgrid')

# x: Competencia del sistema (NEP) | y: Fuerza del triunfo
# hue: Identidad política del ganador
# s=150 hace los puntos más grandes para ver mejor la distribución
sns.relplot(x = 'NEP', 
            y = 'Porcentaje partido ganador', 
            data = df_elecciones,  
            hue = 'Partido ganador',
            s = 150,
            height = 5, 
            aspect = 1.5)
# Elegimos algunas provincias que queremos que tengan la etiqueta de texto - OPCIONAL
provincias_a_marcar = ['BUENOS AIRES', 'CORDOBA', 'SANTA FE', 'MENDOZA', 'FORMOSA']
for _, fila in df_elecciones.iterrows():
    if fila['Provincia'] in provincias_a_marcar:
        plt.text(fila['NEP'], fila['Porcentaje partido ganador'], fila['Provincia'], fontsize=9, fontweight='bold')        
        plt.legend(bbox_to_anchor=(1.02, 1))
plt.title('Relación entre Competencia del Sistema (NEP) y Porcentaje de Victoria')
plt.show()
```

### LINEPLOT
+ Es fundamental para el análisis de series temporales. Sirve para graficar la evolución de la imagen de un dirigente en encuestas mensuales o la variación histórica del ausentismo electoral en diferentes periodos democráticos.
+ También nos sirve para hacer analisis de variación no necesariamente temporal.

```python
# Para generar graficos de serie temporal, usamos otro dataset con comparativa de elecciones presidenciales.
# 1. Cargamos los datos
df_comparativa_elecciones = pd.read_csv('media/datos_electorales_presidente - comparativa temporal.csv')
# 2. LIMPIEZA CLAVE: Nos quedamos con 1 sola fila por provincia/año para no falsear la estadística
df_lineas = df_comparativa_elecciones.drop_duplicates(subset=['id', 'Elecciones', 'Participacion']).copy()

# Ordenamos cronológicamente para que la línea no vaya de adelante hacia atrás
df_lineas['Elecciones'] = pd.Categorical(df_lineas['Elecciones'], 
                                         categories=['GENERALES 2011', 'GENERALES 2015', 'GENERALES 2019', 'GENERALES 2023'], 
                                         ordered=True)
df_lineas = df_lineas.sort_values('Elecciones')

# Elegimos 4 provincias para los gráficos con 'hue' (24 líneas serían ilegibles)
provincias_elegidas = ['BUENOS AIRES', 'CORDOBA', 'SANTA FE', 'MENDOZA'] 
df_4provincias = df_lineas[df_lineas['id'].isin(provincias_elegidas)]
```

```python
# 1. RELPLOT: Línea con Intervalo de Confianza (sombra)
sns.relplot(x='Elecciones', y='Participacion', kind='line', data=df_lineas, height=4, aspect=1.5)
plt.title('1. Evolución de la Participación (Promedio Nacional + Sombra de Confianza)')
plt.xticks(rotation=15)
plt.show()
```

```python
# ci=False apaga el sombreado del intervalo de confianza para ver la línea limpia
sns.relplot(x='Elecciones', y='Participacion', kind='line', data=df_lineas, errorbar=None, height=4, aspect=1.5)
plt.title('2. Evolución de la Participación (Solo Línea Promedio)')
plt.xticks(rotation=15)
plt.show()
```

```python
# Agregamos hue para ver una línea de tendencia por cada fuerza política
sns.relplot(x='Elecciones', y='Participacion', kind='line', hue='id', data=df_4provincias, height=4, aspect=1.5)
plt.title('3. Participación comparada por Provincia Principal')
plt.xticks(rotation=15)
plt.show()
```

```python
# LINEPLOT: en lugar de especificar el kind='line' introducimos la funcion sns.lineplot()
plt.figure(figsize=(8, 4))
sns.lineplot(x='Elecciones', y='Participacion', data=df_lineas)
plt.title('4. Lineplot simple (Promedio Nacional)')
plt.xticks(rotation=15)
plt.show()
```

```python
# LINEPLOT: Color y Estilo por provincia
plt.figure(figsize=(8, 4))
sns.lineplot(x='Elecciones', y='Participacion', hue='id', style='id', data=df_4provincias)
plt.title('5. Lineplot con Color y Tipo de Línea')
plt.xticks(rotation=15)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') # Mueve la leyenda afuera
plt.show()
```

```python
# LINEPLOT: Color, Estilo y Marcadores (Puntitos)
plt.figure(figsize=(8, 4))
sns.lineplot(x='Elecciones', y='Participacion', hue='id', style='id', markers=True, markersize=8, data=df_4provincias)
plt.title('6. Lineplot Completo con Marcadores Exactos')
plt.xticks(rotation=15)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

## Visualización de la distribucion de datos
+ La visualización de distribución consiste en representar gráficamente la frecuencia, forma y dispersión de los valores de una variable cuantitativa dentro de un conjunto de datos. Su objetivo primordial es identificar cómo se reparten las observaciones a lo largo del rango de valores posibles.

### HISTPLOT y DISPLOT
+ El histograma particiona el rango de una variable cuantitativa en intervalos discretos llamados bins (contenedores) y cuenta cuántas observaciones caen en cada uno. 
+ Mientras que histplot() es una función a nivel de ejes, displot() es una función a nivel de figura que permite realizar faceteado (comparar distribuciones en múltiples sub-gráficos).

```python
# histogramas para ver la distribucion de frecuencias
x = df_elecciones['Participacion'].values
# Agregamos kde=True para forzar que dibuje la línea de tendencia sobre las barras
sns.displot(x, color='blue', rug=True, kde=True)
plt.title('Distribución de frecuencias: Participación Electoral')
plt.show()
```

```python
plt.figure(figsize=(8, 5))

# como sacar la media 
x = df_elecciones['Participacion'].values
sns.histplot(x, color='blue', kde=True)

# calculamos la media matemática
mean = df_elecciones['Participacion'].mean()

# especifica que muestre la media como una linea vertical sobre el histplot
plt.axvline(mean, 0, 1, color='red') 

# Agregamos el título (opcional, para que se vea el número exacto en el gráfico)
plt.title(f'Distribución de la Participación (Media: {mean:.2f}%)')
plt.show()
```

### KDEPLOT: kernel density estimation. 
+ Es analoga a los histplot y grafica una curva de densidad continua.
+ A diferencia del histograma, que es discreto, el KDE es una estimación no paramétrica de la función de densidad de probabilidad.
+ El parámetro clave es el bandwidth (ancho de banda): un valor bajo produce una curva ruidosa (overfitting), mientras que un valor alto suaviza demasiado la realidad (underfitting).

```python

plt.figure(figsize=(8, 5))

# 1) kdeplot con sombreado bajo la curva
x = df_elecciones['Porcentaje partido ganador'].values
sns.kdeplot(x, fill=True, color='purple')

plt.title('Densidad del Porcentaje de Victoria')
plt.show()
```

```python
# 2) Ajustamos la suavidad de la curva y cortamos en los límites reales de los datos
ax = sns.kdeplot(data=df_elecciones, x="Porcentaje partido ganador", bw_adjust=5, cut=0, color='green')

plt.title('Tendencia Suavizada del Porcentaje de Victoria (bw_adjust=5, cut=0)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# 3) Curvas apiladas usando una variable categórica (Región)
sns.kdeplot(data=df_elecciones, x="Participacion", hue="Region", multiple="stack")

plt.title('Distribución de la Participación Electoral Apilada por Región')
plt.show()
```

```python
#Utilizamos el dataset de elecciones comparadas
#Cuando el hue es numerico
plt.figure(figsize=(10, 6))

sns.kdeplot(data=df_lineas, 
            x="Participacion", 
            hue="Año", 
            palette="viridis", # El gradiente de colores (ej: claro es 2011, oscuro es 2023)
            linewidth=2,
            warn_singular=False) # Evita warnings si algún año tiene poca dispersión

plt.title('Densidad de la Participación Electoral (2011-2023)', fontsize=14)
plt.xlabel('Participación Provincial (%)', fontsize=12)
plt.ylabel('Densidad', fontsize=12)
plt.show()
df_lineas['Año'] = df_lineas['Elecciones'].astype(str).str.extract(r'(\d+)').astype(int)
```

## Regresiones lineales 

+ Regplot () se puede considerar como un subconjunto de las características de lmplot ()
+ La diferencia entre regplot () y lmplot () es que cuando desea agregar una tercera o más variables, solo puede usar la última; porque lmplot () combina regplot () y FacetGrid



```python

# 1) Regresión lineal: relación entre competitividad del sistema y fuerza del ganador
ax = sns.lmplot(x="NEP", 
                y="Porcentaje partido ganador", 
                data=df_elecciones, 
                height=9)

plt.title('Regresión Lineal: NEP vs Porcentaje del Ganador')
plt.show()
# Hipótesis: A medida que sube el Número Efectivo de Partidos (X), el porcentaje que logra el ganador (Y) debería bajar.
```

```python
# 1. Filtramos el dataset para quedarnos solo con las fuerzas que tienen suficientes datos
df_filtrado = df_elecciones[df_elecciones['Partido ganador'] != 'JUNTOS POR EL CAMBIO']
# 2 Separamos las regresiones en columnas según la variable categórica
ax = sns.lmplot(x="NEP", 
                y="Porcentaje partido ganador", 
                data=df_filtrado, 
                col='Partido ganador',hue='Partido ganador')
```

```python
# Usamos el mismo df_filtrado y ahora sí coinciden los 2 marcadores con los 2 partidos
g = sns.lmplot(x="NEP", 
               y="Porcentaje partido ganador", # Cambiamos a victoria
               hue="Partido ganador",          # El color sigue al ganador (UP o LLA)
               truncate=False, 
               markers=["o", "x"], 
               data=df_filtrado,
               height=6)

plt.title('Regresión Múltiple: NEP vs Victoria (Solo LLA y UP)')
plt.show()
```

```python
# Hipótesis: Relación entre Ausentismo y Voto en Blanco
sns.regplot(x='Participacion', 
            y='Porcentaje voto blanco', 
            data=df_elecciones,
            scatter_kws={'color': 'darkblue', 'alpha': 0.6},
            line_kws={'color': 'orange'})
```

## Grillas multiples
+ Para el análisis exploratorio multivariado de bases de datos grandes como las electorales.
+ Fundamental para crear Matrices de Correlación.

```python
fig, axes = plt.subplots(2, 2, figsize=(13, 8))

# plot a (Arriba a la izquierda) - Concentración
sns.histplot(df_elecciones['Concentración'], color='blue', ax=axes[0,0])
axes[0,0].set_title('Distribución de Concentración')

# plot b (Arriba a la derecha) - NEP
sns.histplot(df_elecciones['NEP'], color='blue', ax=axes[0,1])
axes[0,1].set_title('Distribución del NEP')

# plot c (Abajo a la izquierda) - Participación
sns.histplot(df_elecciones['Participacion'], color='blue', ax=axes[1,0])
axes[1,0].set_title('Distribución de Participación')

# plot d (Abajo a la derecha) - Boxplot por Partido
sns.boxplot(x="2do Partido", y="Porcentaje 2do partido", data=df_elecciones, color='green', ax=axes[1,1])
axes[1,1].set_title('Porcentaje de la 2da Fuerza según su Identidad Política')
axes[1,1].tick_params(axis='x', rotation=15)

plt.tight_layout() 
plt.show()
```

```python
#esta función creará una cuadrícula de ejes: y en una sola fila y los ejes x en una sola columna. 
#diagonales: se dibuja una gráfica de distribución univariante para mostrar la distribución marginal de los datos en cada columna
columnas_numericas = ['Concentración', 'NEP', 'Participacion', 'Porcentaje partido ganador']
g1 = sns.pairplot(df_elecciones[columnas_numericas], 
                  plot_kws={'color':'red'})
plt.show()
```

```python
g2 = sns.pairplot(df_elecciones[columnas_numericas + ['Partido ganador']], 
                  hue='Partido ganador')
plt.show()
```

```python
g3 = sns.pairplot(df_elecciones[columnas_numericas + ['Partido ganador']], 
                  hue='Partido ganador', 
                  kind='reg')
plt.show()
```

```python
# 1. Jointplot con regresión
g = sns.jointplot(x="NEP", 
                  y="Porcentaje partido ganador", 
                  data=df_elecciones, 
                  kind="reg", 
                  color='green', 
                  height=8) # 'height' reemplaza al antiguo 'size'

g.fig.suptitle('Relación NEP vs Fuerza del Triunfo', y=1.03)
plt.show()
```

```python
# 2. Jointplot de densidad (KDE)
g = sns.jointplot(x="Concentración", 
                  y="Porcentaje partido ganador", 
                  data=df_elecciones, 
                  kind="kde", 
                  color='green', 
                  fill=True, 
                  height=8)

g.fig.suptitle('Densidad: Concentración vs Fuerza del Triunfo', y=1.03)
plt.show()
```

## Matrix/colores

```python
#Trazar datos rectangulares como una matriz codificada por colores
# Queremos ver cómo interactúan los conceptos:
variables_matriz = [
    'NEP', 
    'Concentración', 
    'Participacion', 
    'Porcentaje partido ganador', 
    'Porcentaje voto blanco'
]
# 2. Calculamos la correlación de Pearson
matriz_corr = df_elecciones[variables_matriz].corr()
# 3. Creamos el Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(matriz_corr, 
            annot=True,       # Ponemos los números adentro
            cmap='RdBu_r',    # Escala Rojo (positivo) - Azul (negativo)
            center=0,         # El 0 es el color neutro
            fmt=".2f", 
            linewidths=0.5)
plt.title('Matriz Explicativa: La Mecánica del Sistema de Partidos 2023', fontsize=15)
plt.show()
```

```python
# Otros heatmaps
# Usamos el dataset con datos de las elecciones de 2019 en CABA
df_2019CABA = pd.read_csv('media/datos_agrup.csv')
#Acomodamos y limpiamos
df_2019CABA = (
    df_2019CABA
    .loc[df_2019CABA.CODIGO_CATEGORIA == 100000000000]
    .filter(["NOMBRE_AGRUPACION", "VOTOS_AGRUPACION",
             "NOMBRE_REGION","CODIGO_CIRCUITO","CODIGO_MESA"])
    .pivot_table(
        values='VOTOS_AGRUPACION', 
        index=['NOMBRE_REGION',"CODIGO_CIRCUITO","CODIGO_MESA"], 
        columns=['NOMBRE_AGRUPACION'])
)

df_2019CABA.reset_index(level=['NOMBRE_REGION',"CODIGO_CIRCUITO","CODIGO_MESA"], inplace=True)
df_2019CABA = df_2019CABA.loc[(df_2019CABA["JUNTOS POR EL CAMBIO"] + df_2019CABA["JUNTOS POR EL CAMBIO"] )>0]

df_2019CABA = df_2019CABA.rename_axis('index')

df_2019CABA.head()
```

```python
# apply(..., 1) (axis=1 / Horizontal / Por Filas):
    # Las filas suman 100% (1.0). Es ideal para ver la distribución territorial de una fuerza.
sns.heatmap(df_2019CABA
            .filter([ "JUNTOS POR EL CAMBIO","FRENTE DE TODOS","NOMBRE_REGION"])
            .groupby("NOMBRE_REGION")
            .sum()
            .apply(lambda x: (x+0.)/sum(x),1)
            , annot=True
            ,cmap="YlGnBu"
)
```

```python
# apply(..., 0) (axis=0 / Vertical / Por Columnas):
    # Las columnas suman 100% (1.0). Es ideal para ver la distribución territorial de una fuerza.
    # Pregunta que responde: "Del 100% de los votos que sacó ESTE PARTIDO a nivel nacional, ¿cuánto se concentró en cada región?".
sns.heatmap(df_2019CABA
            .filter([ "JUNTOS POR EL CAMBIO","FRENTE DE TODOS","NOMBRE_REGION"])
            .groupby("NOMBRE_REGION")
            .sum()
            .apply(lambda x: (x+0.)/sum(x),0)
            , annot=True
            ,cmap="YlGnBu"
)
```
